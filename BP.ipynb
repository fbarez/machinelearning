```python
f = lambda x: 1.0/(1.0 + np.exp(-x)) # activation function (sigmiod)

X = np.random.rand(3, 1) #random vector of 3 numbers (3x1)
W1, W2, W3 = 0.4, 0.5, 0.1
b1, b2, b3 = 1, 1, 1
h1 = f(np.dot(W1, X) + b1) # calculate first hidden layer activation (4x1)
h2 = f(np.dot(W2, h1) + b2) # calculate second hidden layer acivation

out = np.dot(W3, h2) + b3 #output neuron (1x1)
```

```python
from numpy.random import rand

N, D_in, H, D_out = 64, 1000, 100, 10

x, y = rand(N, D_in), rand(N, D_out)
w1, w2 = rand(D_in, H), rand(H, D_out)

for t in range(2000):
    h = 1/(1+np.exp(-x.dot(w1)))
    y_pred = h.dot(w2)
    loss = np.square(y_pred, -y).sum()
    print(t, loss)
```

```python
grad_y_pred = 2.0 * (y_pred - y)
grad_w2 = h.T.dot(grad_y_pred)
grad_h = grad_y_pred.dot(w2.T)
grad_w1 = x.T.dot(grad_h * h * (1-h))

w1 -= 1e-4 * grad_w1
w2 -= 1e-4 * grad_w2
```
